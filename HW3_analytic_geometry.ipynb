Напишите код, который будет переводить полярные координаты в декартовы.



```python
rads = np.linspace(0,100,10)
degs = np.linspace(0,math.pi,10)
XY = []

for i, j in zip(rads, degs):
    XY.append((i*np.cos(math.pi*j/180), i*np.sin(math.pi*j/180)))
    
XY
```


###4. Задание

Решите систему уравнений:

exp(x) + x(1 – y) = 1

y = x2 – 1


```python
from scipy.optimize import fsolve
import math

def eqs(p):
    x, y = p
    return (np.exp(x) + x*(1 - y) - 1, y - x**2 + 1)

x1, y1 =  fsolve(eqs, (-5, 1))
x2, y2 =  fsolve(eqs, (5, 1))

print (x1, y1)
print (x2, y2)
```
